### Otsu's Binarization

Algorithm too determine the threshold automatically with Linear Discriminant Analysis (LDA)

```python
im = cv2.imread('bimodal_sample.png')
# plt.imshow(im)
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
gray.shape
```

```python
sns.set('notebook', 'ticks')
sns.distplot(gray)
# plt.vlines(retval, 0, 0.01, 'r')
```

```python
retval, dst = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(dst, cmap='gray')
```

### cv2.adaptiveThreshold()

```python
im = plt.imread('text_pic.jpg')
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
dst = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 4)
plt.imshow(dst, cmap='gray')
```
